```python
# Data 
import feedparser
from newspaper import Article

# Analysis
import tensorflow_hub as hub
import numpy as np

# Misc 
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
```

```python
threatpost = "https://threatpost.com/feed/"
securityweek = "http://feeds.feedburner.com/Securityweek?format=xml"
```

```python
# Read outlets
threatpost_posts = feedparser.parse(threatpost)
securityweek_posts = feedparser.parse(securityweek)

# Concat
all_posts = threatpost_posts["entries"] + securityweek_posts["entries"]
```

```python
for post in tqdm(all_posts):
    url = 'http://fox13now.com/2013/12/30/new-year-new-laws-obamacare-pot-guns-and-drones/'
    article = Article(post["link"])
    article.download()
    article.parse()
    post["article_text"] = article.text
```

```python
embeddings = embed([p["article_text"] for p in all_posts])
```

```python
similarities = np.inner(embeddings, embeddings)
```

```python
plt.hist(similarities.flatten())
```

```python
for idx, post in enumerate(all_posts):
    similarity_by_index = np.argsort(similarities[idx])[::-1]
    
    first = True
    
    for sidx in similarity_by_index[1:]:
               
        sim = similarities[idx][sidx]
        if sim < 0.75:
            continue
            
        if first:
            first = False
            print("\n")
            print(post["title"])
    
        post_sim = all_posts[sidx]
        print("--({})-> {}".format(np.round(sim, 2), post_sim["title"]))
```

```python
plt.figure(figsize=(20, 10))
sms = plt.matshow(similarities, fignum=1)
plt.colorbar(sms)
```
